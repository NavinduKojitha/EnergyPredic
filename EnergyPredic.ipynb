```python
# Appliance Energy Prediction Using Deep Learning
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.feature_selection import RFE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class EnergyPredictionSystem:
    """
    Complete system for appliance energy prediction using deep learning
    """

    def __init__(self):
        self.scaler = StandardScaler()
        self.model = None
        self.baseline_models = {}
        self.feature_names = []
        self.results = {}

    def load_and_explore_data(self, file_path=None):
        """
        Load and perform initial exploration of the dataset
        """
        print("Loading dataset...")
        from google.colab import files
        uploaded = files.upload()
        self.df = pd.read_csv(next(iter(uploaded)))

        print(f"Dataset loaded successfully!")
        print(f"Shape: {self.df.shape}")

# Convert date column if it exists
        if 'date' in self.df.columns:
            self.df['date'] = pd.to_datetime(self.df['date'])
            print(f"Date range: {self.df['date'].min()} to {self.df['date'].max()}")

        return self.df

    def exploratory_data_analysis(self):
        """
        Perform comprehensive exploratory data analysis
        """
        print("\n=== EXPLORATORY DATA ANALYSIS ===")

# Basic statistics
        print("\nDataset Info:")
        print(self.df.info())
        print("\nBasic Statistics:")
        print(self.df.describe())

# Check missing values
        print("\nMissing Values:")
        missing_values = self.df.isnull().sum()
        print(missing_values[missing_values > 0])

# Create visualizations
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Time series plot of target variable if date column exists
        if 'date' in self.df.columns and 'Appliances' in self.df.columns:
            axes[0, 0].plot(self.df['date'], self.df['Appliances'])
            axes[0, 0].set_title('Appliance Energy Consumption Over Time')
            axes[0, 0].set_xlabel('Date')
            axes[0, 0].set_ylabel('Energy (Wh)')
            axes[0, 0].tick_params(axis='x', rotation=45)
        else:
            axes[0, 0].text(0.5, 0.5, 'No date/Appliances columns for time series plot',
                           ha='center', va='center')
            axes[0, 0].set_title('Time Series Data Not Available')

# Distribution of target variable
        if 'Appliances' in self.df.columns:
            axes[0, 1].hist(self.df['Appliances'], bins=50, alpha=0.7)
            axes[0, 1].set_title('Distribution of Appliance Energy Consumption')
            axes[0, 1].set_xlabel('Energy (Wh)')
            axes[0, 1].set_ylabel('Frequency')
        else:
            axes[0, 1].text(0.5, 0.5, 'No Appliances column for distribution plot',
                           ha='center', va='center')
            axes[0, 1].set_title('Target Variable Not Found')

# Correlation heatmap
        numeric_cols = self.df.select_dtypes(include=[np.number]).columns
        if len(numeric_cols) > 0:
            correlation_matrix = self.df[numeric_cols].corr()
            im = axes[1, 0].imshow(correlation_matrix, cmap='coolwarm', aspect='auto')
            axes[1, 0].set_title('Correlation Matrix')
            axes[1, 0].set_xticks(range(len(numeric_cols)))
            axes[1, 0].set_yticks(range(len(numeric_cols)))
            axes[1, 0].set_xticklabels(numeric_cols, rotation=45)
            axes[1, 0].set_yticklabels(numeric_cols)
        else:
            axes[1, 0].text(0.5, 0.5, 'No numeric columns for correlation matrix',
                           ha='center', va='center')
            axes[1, 0].set_title('No Numeric Data')

# Seasonal patterns if date and target exist
        if 'date' in self.df.columns and 'Appliances' in self.df.columns:
            self.df['hour'] = self.df['date'].dt.hour
            hourly_avg = self.df.groupby('hour')['Appliances'].mean()
            axes[1, 1].plot(hourly_avg.index, hourly_avg.values)
            axes[1, 1].set_title('Average Energy Consumption by Hour')
            axes[1, 1].set_xlabel('Hour of Day')
            axes[1, 1].set_ylabel('Average Energy (Wh)')
        else:
            axes[1, 1].text(0.5, 0.5, 'No date/Appliances columns for seasonal analysis',
                           ha='center', va='center')
            axes[1, 1].set_title('Seasonal Analysis Not Available')

        plt.tight_layout()
        plt.show()

#analysis
        if len(numeric_cols) > 0 and 'Appliances' in numeric_cols:
            print("\nCorrelation with target variable:")
            target_corr = correlation_matrix['Appliances'].sort_values(ascending=False)
            print(target_corr)
            return correlation_matrix
        else:
            print("\nNo numeric target variable for correlation analysis")
            return None

    def preprocess_data(self):
        """
        Handle missing values, outliers, and scaling
        """
        print("\n=== DATA PREPROCESSING ===")

# Handle missing values
        if self.df.isnull().sum().sum() > 0:
            print("Handling missing values...")
# Forward fill for time series data
            self.df = self.df.fillna(method='ffill').fillna(method='bfill')

# Outlier detection and IQR method
        print("Detecting and treating outliers...")
        numeric_columns = self.df.select_dtypes(include=[np.number]).columns

        for col in numeric_columns:
            if 'Appliances' not in col:
                Q1 = self.df[col].quantile(0.25)
                Q3 = self.df[col].quantile(0.75)
                IQR = Q3 - Q1
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

#  removing outliers
                self.df[col] = np.clip(self.df[col], lower_bound, upper_bound)

# Encode variables
        print("Encoding categorical variables...")
        if 'WeekStatus' in self.df.columns:
            self.df['WeekStatus_Weekend'] = (self.df['WeekStatus'] == 'Weekend').astype(int)

        print("Preprocessing completed!")

    def feature_engineering(self):
        """
        Create and select features for model training
        """
        print("\n=== FEATURE ENGINEERING ===")

# Time-based features
        if 'date' in self.df.columns:
            print("Creating time-based features...")
            self.df['hour'] = self.df['date'].dt.hour
            self.df['day'] = self.df['date'].dt.day
            self.df['month'] = self.df['date'].dt.month
            self.df['year'] = self.df['date'].dt.year
            self.df['dayofyear'] = self.df['date'].dt.dayofyear
            self.df['weekofyear'] = self.df['date'].dt.isocalendar().week

# Cyclical encoding
            self.df['hour_sin'] = np.sin(2 * np.pi * self.df['hour'] / 24)
            self.df['hour_cos'] = np.cos(2 * np.pi * self.df['hour'] / 24)
            if 'Day_of_week' in self.df.columns:
                self.df['day_sin'] = np.sin(2 * np.pi * self.df['Day_of_week'] / 7)
                self.df['day_cos'] = np.cos(2 * np.pi * self.df['Day_of_week'] / 7)
            self.df['month_sin'] = np.sin(2 * np.pi * self.df['month'] / 12)
            self.df['month_cos'] = np.cos(2 * np.pi * self.df['month'] / 12)

# Rolling averages
        if 'Appliances' in self.df.columns:
            print("Creating rolling averages...")
            windows = [6, 18, 144]
            for window in windows:
                self.df[f'Appliances_rolling_{window}'] = self.df['Appliances'].rolling(window=window, min_periods=1).mean()
                if 'T1' in self.df.columns:
                    self.df[f'T1_rolling_{window}'] = self.df['T1'].rolling(window=window, min_periods=1).mean()
                if 'RH_1' in self.df.columns:
                    self.df[f'RH_1_rolling_{window}'] = self.df['RH_1'].rolling(window=window, min_periods=1).mean()

# Lagged features
        print("Creating lagged features...")
        lags = [1, 6, 18, 144]
        for lag in lags:
            if 'Appliances' in self.df.columns:
                self.df[f'Appliances_lag_{lag}'] = self.df['Appliances'].shift(lag)
            if 'lights' in self.df.columns:
                self.df[f'lights_lag_{lag}'] = self.df['lights'].shift(lag)

# Interaction features
        print("Creating interaction features...")
        if 'T1' in self.df.columns and 'RH_1' in self.df.columns:
            self.df['T1_RH_1_interaction'] = self.df['T1'] * self.df['RH_1']
        if 'T_out' in self.df.columns and 'RH_out' in self.df.columns:
            self.df['T_out_RH_out_interaction'] = self.df['T_out'] * self.df['RH_out']
        if 'T1' in self.df.columns and 'T_out' in self.df.columns:
            self.df['temp_diff'] = self.df['T1'] - self.df['T_out']
        if 'RH_1' in self.df.columns and 'RH_out' in self.df.columns:
            self.df['humidity_diff'] = self.df['RH_1'] - self.df['RH_out']

# Statistical features
        temp_cols = [col for col in self.df.columns if col.startswith('T') and col[1:].isdigit()]
        if temp_cols:
            self.df['temp_avg'] = self.df[temp_cols].mean(axis=1)
            self.df['temp_std'] = self.df[temp_cols].std(axis=1)

        humidity_cols = [col for col in self.df.columns if col.startswith('RH_') and col[3:].isdigit()]
        if humidity_cols:
            self.df['humidity_avg'] = self.df[humidity_cols].mean(axis=1)
            self.df['humidity_std'] = self.df[humidity_cols].std(axis=1)

# Drop rows with NaN values
        self.df = self.df.dropna()

        print(f"Feature engineering completed! New shape: {self.df.shape}")

    def prepare_features(self):
        """
        Select and prepare final feature set
        """
        print("\n=== FEATURE SELECTION ===")

# target variable
        if 'Appliances' not in self.df.columns:
            raise ValueError("Target variable 'Appliances' not found in dataset")

# Define feature columns
        exclude_cols = ['date', 'Appliances', 'WeekStatus']
        feature_columns = [col for col in self.df.columns if col not in exclude_cols]

        X = self.df[feature_columns]
        y = self.df['Appliances']

# Feature selection using Random Forest
        rf_selector = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_selector.fit(X, y)

# feature importance
        feature_importance = pd.DataFrame({
            'feature': feature_columns,
            'importance': rf_selector.feature_importances_
        }).sort_values('importance', ascending=False)

        print("Top 20 Most Important Features:")
        print(feature_importance.head(20))

# Select top features
        top_features = feature_importance.head(30)['feature'].tolist()
        self.feature_names = top_features

        return X[top_features], y

    def split_data(self, X, y, test_size=0.2):
        """
        Split data maintaining temporal order
        """
        print(f"\n=== DATA SPLITTING ===")

# data leakage
        split_index = int(len(X) * (1 - test_size))

        X_train = X.iloc[:split_index]
        X_test = X.iloc[split_index:]
        y_train = y.iloc[:split_index]
        y_test = y.iloc[split_index:]

        print(f"Training set: {X_train.shape}")
        print(f"Test set: {X_test.shape}")

        return X_train, X_test, y_train, y_test

    def scale_features(self, X_train, X_test):
        """
        Apply feature scaling
        """
        print("Scaling features...")

        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)

        return X_train_scaled, X_test_scaled

    def build_baseline_models(self, X_train, X_test, y_train, y_test):
        """
        Build baseline models for comparison
        """
        print("\n=== BASELINE MODELS ===")

# Linear Regression
        lr = LinearRegression()
        lr.fit(X_train, y_train)
        lr_pred = lr.predict(X_test)

# Random Forest
        rf = RandomForestRegressor(n_estimators=100, random_state=42)
        rf.fit(X_train, y_train)
        rf_pred = rf.predict(X_test)

# Store baseline models and predictions
        self.baseline_models = {
            'Linear Regression': {'model': lr, 'predictions': lr_pred},
            'Random Forest': {'model': rf, 'predictions': rf_pred}
        }

# Evaluate baseline models
        for name, model_data in self.baseline_models.items():
            pred = model_data['predictions']
            mae = mean_absolute_error(y_test, pred)
            rmse = np.sqrt(mean_squared_error(y_test, pred))
            r2 = r2_score(y_test, pred)

            print(f"{name} - MAE: {mae:.2f}, RMSE: {rmse:.2f}, R2: {r2:.3f}")

    def create_sequences(self, X, y, sequence_length=24):
        """
        Create sequences for LSTM model
        """
# Convert to numpy arrays
        X = X.values if hasattr(X, 'values') else X
        y = y.values if hasattr(y, 'values') else y

        X_seq, y_seq = [], []

        for i in range(sequence_length, len(X)):
            X_seq.append(X[i-sequence_length:i])
            y_seq.append(y[i])

        return np.array(X_seq), np.array(y_seq)

    def build_deep_learning_model(self, input_shape):
        """
        Build and compile deep learning model
        """
        print("\n=== DEEP LEARNING MODEL ===")

        model = Sequential([
# LSTM layers
            LSTM(100, return_sequences=True, input_shape=input_shape),
            Dropout(0.2),
            LSTM(50, return_sequences=False),
            Dropout(0.2),

# Dense layers
            Dense(50, activation='relu'),
            Dropout(0.2),
            Dense(25, activation='relu'),
            Dense(1)
        ])

# Compile model
        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='mse',
            metrics=['mae']
        )

        print("Model Architecture:")
        model.summary()

        return model

    def train_deep_learning_model(self, X_train, X_test, y_train, y_test):
        """
        Train the deep learning model
        """
        print("Training deep learning model...")

# Create sequences for LSTM
        sequence_length = 24
        X_train_seq, y_train_seq = self.create_sequences(X_train, y_train, sequence_length)
        X_test_seq, y_test_seq = self.create_sequences(X_test, y_test, sequence_length)

        print(f"Sequence shapes - Train: {X_train_seq.shape}, Test: {X_test_seq.shape}")

# Build model
        self.model = self.build_deep_learning_model((sequence_length, X_train.shape[1]))

# Callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        )

        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7
        )

# Train model
        history = self.model.fit(
            X_train_seq, y_train_seq,
            batch_size=32,
            epochs=50,
            validation_data=(X_test_seq, y_test_seq),
            callbacks=[early_stopping, reduce_lr],
            verbose=1
        )

# Make predictions
        y_pred = self.model.predict(X_test_seq)

        return y_test_seq, y_pred.flatten(), history

    def evaluate_models(self, y_test, y_pred_dl):
        """
        Evaluate all models and compare performance
        """
        print("\n=== MODEL EVALUATION ===")

# Deep Learning model evaluation
        mae_dl = mean_absolute_error(y_test, y_pred_dl)
        rmse_dl = np.sqrt(mean_squared_error(y_test, y_pred_dl))
        r2_dl = r2_score(y_test, y_pred_dl)

        print(f"Deep Learning Model:")
        print(f"  MAE: {mae_dl:.2f}")
        print(f"  RMSE: {rmse_dl:.2f}")
        print(f"  R2: {r2_dl:.3f}")

# results
        self.results['Deep Learning'] = {
            'MAE': mae_dl,
            'RMSE': rmse_dl,
            'R2': r2_dl,
            'predictions': y_pred_dl
        }

# Compare with baselines
        if hasattr(self, 'baseline_models'):
            print("\nBaseline Comparisons:")
            for name, model_data in self.baseline_models.items():
                pred = model_data['predictions']

                if len(pred) > len(y_test):
                    pred = pred[-len(y_test):]
                    y_test_adj = y_test
                elif len(pred) < len(y_test):
                    y_test_adj = y_test[-len(pred):]
                else:
                    y_test_adj = y_test

                mae = mean_absolute_error(y_test_adj, pred)
                rmse = np.sqrt(mean_squared_error(y_test_adj, pred))
                r2 = r2_score(y_test_adj, pred)

# Store baseline results
                self.results[name] = {
                    'MAE': mae,
                    'RMSE': rmse,
                    'R2': r2,
                    'predictions': pred
                }

                print(f"  {name} - MAE: {mae:.2f}, RMSE: {rmse:.2f}, R2: {r2:.3f}")

    def visualize_results(self, y_test, y_pred, history=None):
        """
        Create visualizations of model performance
        """
        print("\n=== VISUALIZATION ===")

        fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Predicted vs Actual
        axes[0, 0].scatter(y_test, y_pred, alpha=0.5)
        axes[0, 0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
        axes[0, 0].set_xlabel('Actual Values')
        axes[0, 0].set_ylabel('Predicted Values')
        axes[0, 0].set_title('Predicted vs Actual Values')

# Residuals plot
        residuals = y_test - y_pred
        axes[0, 1].scatter(y_pred, residuals, alpha=0.5)
        axes[0, 1].axhline(y=0, color='r', linestyle='--')
        axes[0, 1].set_xlabel('Predicted Values')
        axes[0, 1].set_ylabel('Residuals')
        axes[0, 1].set_title('Residual Plot')

# Time series comparison
        n_points = min(500, len(y_test))
        axes[1, 0].plot(range(n_points), y_test[-n_points:], label='Actual', alpha=0.7)
        axes[1, 0].plot(range(n_points), y_pred[-n_points:], label='Predicted', alpha=0.7)
        axes[1, 0].set_xlabel('Time Steps')
        axes[1, 0].set_ylabel('Energy Consumption')
        axes[1, 0].set_title('Time Series Comparison (Last 500 points)')
        axes[1, 0].legend()

# Training history
        if history is not None:
            axes[1, 1].plot(history.history['loss'], label='Training Loss')
            axes[1, 1].plot(history.history['val_loss'], label='Validation Loss')
            axes[1, 1].set_xlabel('Epoch')
            axes[1, 1].set_ylabel('Loss')
            axes[1, 1].set_title('Training History')
            axes[1, 1].legend()

        plt.tight_layout()
        plt.show()

    def run_complete_pipeline(self, file_path=None):
        """
        Run the complete machine learning pipeline
        """
        print("=== APPLIANCE ENERGY PREDICTION SYSTEM ===")
        print("Starting complete pipeline...")

#Load and explore data
        self.load_and_explore_data(file_path)
        correlation_matrix = self.exploratory_data_analysis()

# Preprocess data
        self.preprocess_data()

# Feature engineering
        self.feature_engineering()

# Prepare features
        X, y = self.prepare_features()

# Split data
        X_train, X_test, y_train, y_test = self.split_data(X, y)

# Scale features
        X_train_scaled, X_test_scaled = self.scale_features(X_train, X_test)

#  Build baseline models
        self.build_baseline_models(X_train_scaled, X_test_scaled, y_train, y_test)

# Train deep learning model
        y_test_seq, y_pred_dl, history = self.train_deep_learning_model(
            X_train_scaled, X_test_scaled, y_train, y_test
        )

# Evaluate models
        self.evaluate_models(y_test_seq, y_pred_dl)

# Visualize results
        self.visualize_results(y_test_seq, y_pred_dl, history)

        print("\n=== PIPELINE COMPLETED ===")
        print("All tasks completed successfully!")

        return self.results

# usage and testing
if __name__ == "__main__":

    energy_system = EnergyPredictionSystem()

# Run the complete pipeline
    results = energy_system.run_complete_pipeline()

# Print final results summary
    print("\n=== FINAL RESULTS SUMMARY ===")
    for model_name, metrics in results.items():
        print(f"{model_name}:")
        print(f"  MAE: {metrics['MAE']:.2f}")
        print(f"  RMSE: {metrics['RMSE']:.2f}")
        print(f"  R: {metrics['R2']:.3f}")
```
